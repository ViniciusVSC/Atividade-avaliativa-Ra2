```python
#Função numero 1
# f(x) = (x**2 - 4)(x**3 - 8x -1)

import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x**2 - 4)*(x**3 - 8*x - 1)

x = np.linspace(-10, 10, 100)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função número 1.(x^2-4)(x^3-8x-1)')
plt.grid(True)
plt.show()


```

$f’(x) = \frac{d}{dx}[(x2-4)(x3-8x-1)]$

$f’(x) = (2x)(x^3-8x-1) + (x2-4)(3x2-8)$

```python
#Função número 3
#f(x)=x2e−2xln(x)

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 10, 100)
y = x**2 * np.exp(-2*x) * np.log(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função f(x)=x^2e^(-2x)ln(x)')
plt.grid(True)
plt.show()
```


Função Número 3
 Regra do produto para encontrar a derivada de f(x) em relação a x:

\begin{aligned}
f'(x) &= \frac{d}{dx} \left( x^2 * e^{-2x} * \ln(x) \right) \\
&= \left( \frac{d}{dx} x^2 \right) * e^{-2x} * \ln(x) + x^2 * \left( \frac{d}{dx} e^{-2x} \right) * \ln(x) + x^2 * e^{-2x} * \left( \frac{d}{dx} \ln(x) \right)
\end{aligned}

Calcular cada uma das derivadas:

\begin{aligned}
\frac{d}{dx} x^2 &= 2x \\
\frac{d}{dx} e^{-2x} &= -2e^{-2x} \\
\frac{d}{dx} \ln(x) &= \frac{1}{x}
\end{aligned}

Substituindo essas derivadas na equação original:

\begin{aligned}
f'(x) &= 2x * e^{-2x} * \ln(x) + x^2 * (-2e^{-2x}) * \ln(x) + x^2 * e^{-2x} * \frac{1}{x} \\
&= 2xe^{-2x}\ln(x) - 2x^2e^{-2x}\ln(x) + xe^{-2x}
\end{aligned}

Temos a derivada de f(x) em relação a x, podemos encontrar a inclinação da reta tangente no ponto x=5 substituindo x=5 na equação acima:

\begin{aligned}
f'(5) &= 2*5*e^{-10}\ln(5) - 2*5^2*e^{-10}\ln(5) + 5*e^{-10} \\
&= -10e^{-10}\ln(5)
\end{aligned}

Portanto, o ângulo de inclinação da reta tangente no ponto x=5 é $-10e^{-10}\ln(5)$.

# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto mínimo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=7\sin(x)-2\cos(x)$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+2y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import plotly.graph_objs as go
import numpy as np


def f(x, y):
    return np.sin(x*y) + np.cos(x + 2*y)


x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)
xv, yv = np.meshgrid(x, y)


z = f(xv, yv)


max_index = np.unravel_index(np.argmax(z), z.shape)
min_index = np.unravel_index(np.argmin(z), z.shape)
max_x = xv[max_index]
max_y = yv[max_index]
min_x = xv[min_index]
min_y = yv[min_index]


surface = go.Surface(x=xv, y=yv, z=z)


max_marker = go.Scatter3d(
    x=[max_x],
    y=[max_y],
    z=[z.max()],
    mode='markers',
    marker=dict(
        size=8,
        color='red'
    ),
    name='Máximo'
)

min_marker = go.Scatter3d(
    x=[min_x],
    y=[min_y],
    z=[z.min()],
    mode='markers',
    marker=dict(
        size=8,
        color='blue'
    ),
    name='Mínimo'
)

fig = go.Figure(data=[surface, max_marker, min_marker])
fig.show()

print("Máximo bolinha vermelha, Mínimo bolinha azul")
```
